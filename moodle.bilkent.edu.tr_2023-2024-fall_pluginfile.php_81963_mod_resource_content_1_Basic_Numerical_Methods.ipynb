## Finite Representation of Functions

In computations, arguments of continuous functions take finite values. Numeric computations require the space to be discretized. 

As an example, an interval of space $[a,b]$  in 1-dimensions, can be represented by $N$ distinct space points. Between these points there exist finite distances, $Δx_i\;{\rm where}\;i=1,\dots, N$. The discretization of space can be uniform or non-uniform depending on the further techniques to be used. In any case, the function on discrete space points,

$$f(x_i){\rm where} \;i=0,\dots, N$$

take discrete values. In any case, the function on discrete space points, 
These discrete values of the function can be represented by $f(x_i)$ or simply by using the index of the discrete pace point, $f_i$.

The accuracy of the numerical computations depends on which method is employed to obtain an approximate analytical form that represents the function locally by using the distinct function values. The numerical method of obtaining an analytical form between the discrete points is called interpolation.
A widely used method is Taylor series expansion around the point  $x_i$.



$$ f(x_i+\Delta x) = f(x_i) + \left . \frac{df(x)}{dx} \right|_{x=x_i} \Delta x + \frac{1}{2!}\left . \frac{d^2f(x)}{dx^2} \right|_{x=x_i} {\Delta x}^2 + \frac{1}{3!}\left . \frac{d^3f(x)}{dx^3} \right|_{x=x_i} {\Delta x}^3 +  \cdots $$

$$ f_{i+1} = f_i + \left . \frac{df}{dx} \right|_{x=x_i} h + \frac{1}{2!}\left . \frac{d^2f}{dx^2} \right|_{x=x_i} {h}^2 + \cdots $$


There exist numerous interpolation techniques to represent functions with high accuracy. All numerical methods are evaluated by considering their accuracy, computational complexity, and computational costs. Taylor series expansion is one of the most convenient methods for simple problems. But for high accuracy more advanced techniques are preferable. These methods will be studied in due course.

## Derivatives

The mathematical definition of the first derivative of a function $f(x)$ is,

$$ \frac{df(x)}{dx} = \lim_{\Delta x\rightarrow 0}\frac{f(x+\Delta x) - f(x)}{\Delta x} $$

The derivative, in discrete space, can be written in terms of the Taylor series ,

$$ f(x_i+\Delta x) = f(x_i) + \left . \frac{df(x)}{dx} \right|_{x=x_i} \Delta x + \mathcal{O}({\Delta x}^2) $$

$$ \left . \frac{df(x)}{dx} \right|_{x=x_i} = \frac{f(x_i+\Delta x) - f(x_i)}{\Delta x} + \mathcal{O}({\Delta x}) $$

$$ \left . \frac{df(x)}{dx} \right|_{x=x_i} = \frac{f(x_i+h) - f(x_i)}{h} + \mathcal{O}({h}) $$

Differentiation rule based on keeping a certain number of terms in a Taylor series also provides an expression for the error. The term that the Taylor series is truncated is the source of the error in differentiation.
In this approach, (Forward difference algorithm) it is seen that there is an error in the order of $\mathcal{O}(\Delta x)$. $\Delta x$ can be reduced to reduce the error. Although shrinking $\Delta x$ is associated with error reduction, it has different difficulties (The precision will be lost through the subtractive cancelation on the left-hand side).

An other derivative approach with the same error sensitivity is the backward difference approach. In this approach, the difference between the function value at $i$ and the function value at $i-1$ is used for derivative calculation.

$$ \left . \frac{df(x)}{dx} \right|_{x=x_i} = \frac{f(x_i-h) - f(x_i)}{h} + \mathcal{O}({h}) $$



### CENTRAL DIFFERENCE


In numerical calculation, the purpose should be to reduce the error by incorporating higher-order terms in the expansion, instead of reducing the step size.

Hence, consider two Taylor series expansions, $f_{i+1}$ and $f_{i-1}$

$$ f_{i+1} = f_i + \left . \frac{df}{dx} \right|_{x=x_i} h + \frac{1}{2!}\left . \frac{d^2f}{dx^2} \right|_{x=x_i} {h}^2 + \frac{1}{3!}\left . \frac{d^3f}{dx^3} \right|_{x=x_i} {h}^3 + \frac{1}{4!}\left . \frac{d^4f}{dx^4} \right|_{x=x_i} {h}^4 +  \mathcal{O}({ h^5})$$

$$ f_{i-1} = f_i - \left . \frac{df}{dx} \right|_{x=x_i} h + \frac{1}{2!}\left . \frac{d^2f}{dx^2} \right|_{x=x_i} {h}^2 - \frac{1}{3!}\left . \frac{d^3f}{dx^3} \right|_{x=x_i} {h}^3 + \frac{1}{4!}\left . \frac{d^4f}{dx^4} \right|_{x=x_i} {h}^4 +  \mathcal{O}({ h^5})$$

As a result of subtracting the two equations side by side
The resulting expression provides a derivative expression that also includes higher-order terms.

$$ \left . \frac{df(x)}{dx} \right|_{x=x_i} =\; \frac{f(x_i+h) - f(x_i-h)}{2 h} + \mathcal{O}({h^2}) $$

Or by reducing the step size to half, $h \rightarrow h/2$,
$$ \left . \frac{df(x)}{dx} \right|_{x=x_i} =\; \frac{f(x_i+h/2) - f(x_i-h/2)}{h} + \mathcal{O}({(\frac{h}{2}^2}) $$

Dispite the fact that $\mathcal{O}(h^4)$ terms cancel, since $\mathcal{O}(h^3)$ term remains, the errors is of the order $\mathcal{O}(h^4)$
### EXTRAPOLATED DIFFERENCE

Hence in order to reduce the error, some technique are devised to cancel more and more higher order terms.

Take two central difference equations, and consider how to eliminate the remaining terms to push the truncation error into higher order terms:


In the central difference derivative method, the first neglected term is of the order $\mathcal{O}(h^3)$. The coefficient of the term causing the error of the order  $\mathcal{O}(h^3)$ is $2\times 1/3! 1/2^3$, and, $2\times 1/3! 1/4^3$  for the step lengths $h/2$ and $h/4$ respectively. Combining two central difference equations, ($D_{cd}(h/2)$ and $D_{cd}(h/4)$) by multiplying appropriate constants, the error is pushed up to the order, $\mathcal{O}(h^5)$, since $\mathcal{O}(h^4)$ terms cancel automatically by the difference operation.


$$ \left . \frac{df(x)}{dx} \right|_{x=x_i} =\; \frac{f(x_i+h/2) - f(x_i-h/2)}{h} + 2 \frac{1}{3!} \frac{d^3f}{dx^3} \frac{(h/2)^3}{h}\mathcal{O}({\frac{h}{2}^4}) $$

$$ \left . \frac{df(x)}{dx} \right|_{x=x_i} =\; \frac{f(x_i+h) - f(x_i-h)}{2 h} +  2 \frac{1}{3!} \frac{d^3f}{dx^3} \frac{(h)^3}{2 h}+ \mathcal{O}(h^4) $$

Terms of $\mathcal{O}(h^3)$ cancel giving,

$$ \left . \frac{df(x)}{dx} \right|_{x=x_i} = \frac{\; 4 \left [ \frac{f(x_i+h/2) - f(x_i-h/2)}{h} \right ] - \left [ \frac{f(x_i+h) - f(x_i-h)}{2 h} \right ]}{3} + \mathcal{O}(h^4) $$

## Partial Derivatives

One can use Taylor series expansion to calculate partial derivatives of multivariable functions:


\begin{equation}
f(x+\Delta x, y) = f(x,y) + \frac{\partial f}{\partial x} \Delta x +  \mathcal{O}(\Delta x^2)
\end{equation}

\begin{equation}
\frac{\partial f}{\partial x} = \frac{f(x+\Delta x , y)-f(x,y)}{\Delta x} + \mathcal{O}(\Delta x^2)
\end{equation}

\begin{equation}
f(x+\Delta x, y) = f(x,y) + \frac{\partial f}{\partial x} \Delta x +\frac{1}{2} \frac{\partial^2 f}{\partial x^2} \Delta x^2 + \mathcal{O}(\Delta x^3)
\end{equation}

\begin{equation}
f(x, y+\Delta y) \;=\; f(x,y) + \frac{\partial f}{\partial y} \Delta y +\frac{1}{2} \frac{\partial^2 f}{\partial y^2} \Delta y^2 + \mathcal{O}(\Delta y^3)
\end{equation}

\begin{equation}
f(x + \Delta x, y+\Delta y) \;= \; f(x,y) + \frac{\partial f}{\partial x} \Delta x + \frac{\partial f}{\partial y} \Delta y +\frac{1}{2}(\frac{\partial^2 f}{\partial x^2} \Delta x^2+ \frac{\partial^2 f}{\partial y^2} \Delta y^2) + \frac{\partial^2f} {\partial x \partial y} \Delta x \Delta y
\end{equation}

\begin{equation}
f(x + \Delta x, y+\Delta y) -f(x+ \Delta x, y) -f(x, y+\Delta y) +f(x, y)  \;= \;  \frac{1}{2} \frac{\partial^2f} {\partial x \partial y} \Delta x \Delta y
\end{equation}

\begin{equation}
 \frac{\partial^2f} {\partial x \partial y}  \;= \; \frac{f(x + \Delta x, y+\Delta y) -f(x+\Delta x, y) -f(x, y+\Delta y) +f(x, y)}{\Delta x \Delta y}
\end{equation}



##  SEARCHING

Finding the roots of a function is one of the most commonly met
problems of numerical analysis. By definition, $x=a$ is a root of $f$,
if $f(a)=0$. However, in numerical applications, it must be understood that the equation usually can not be satisfied exactly due to round-off errors and limited capacity. Therefore, the mathematical definition of a root must be modified. For numerical computations, the condition of having a root
at $x=a$, $f(a) = 0$ is modified such that $ \left |f(a)\right|<\epsilon $, where
$\epsilon$ is a given tolerance suffices. The inequality defines an
interval instead of a point.

### NEWTON–RAPHSON

The Newton–Raphson algorithm finds approximate roots of the equation
$$ f(x) = 0 $$
To use Newton–Raphson algorithm  guess  $x_0$, calculate the function value a step forward, $f(x_0+\Delta x)$.

In terms of Taylor series expansion,

$$ f(x_0 + \Delta x) \sim f(x_0) + \left . \frac{df(x)}{dx} \right |_{x = x_0}(x - x_0) + \mathcal{O}({\Delta x}^2) $$

and expresses a new guess $x$ as a correction $\Delta x$ to the old guess.

If $ f(x_0 + \Delta x) \sim 0$ then the root is reached.


$$  f(x_0) + \left . \frac{df(x)}{dx} \right |_{x = x_0}(x - x_0) = 0\;\; \rightarrow \;\; \Delta x = - \frac{f(x_0)}{\left . \frac{df(x)}{dx} \right |_{x = x_0}}$$


Iteratively the root is reached ,

$$ \displaystyle{ x = x_0 - \frac{f(x_0)}{\left . \frac{df(x)}{dx} \right |_{x = x_0}}}$$

### SECANT ALGORITHM

The Newton–Raphson algorithm requires explicit knowledge of the derivative $df /dx$ at
each value of $x_0$.  Complicated problems, use a numerical approximation to the derivative of the function.



$$  x = x_0 - \frac{f(x_0)}{\left . \frac{df(x)}{dx} \right |_{x = x_0}}\;\rightarrow \; x = x_0 - h \frac{f(x_0)}{f(x_0+h/2) - f(x_0-h/2)}$$

where, $\left . \frac{df(x)}{dx} \right |_{x = x_0} $ is replaced with central difference equation,  $\frac{f(x_0+h/2) - f(x_0-h/2)}{h}$. 

This equation is known as the secant formula. The convergence of the secant method is not as good as the Newton-Rapson method. In numerical applications, the secant method is more practical, and can be used in applications where obtaining the analytical form of the function derivatives is not easy. 

#### Drawbacs of  Newton and Secant Algorithms

Newton and Secant algorithms may converge quickly if the initial point is close to the root. If the initial guess is not close to the root, the steps may become unstable and fail to converge.

### BISECTION ALGORITHM



The bisection algorithm is based on the knowledge that $f(x)$ changes sign in the interval in which a root exists.

The bisection algorithm will always converge to the root by finding progressively smaller and smaller intervals in which zero occurs. Convergence is slow in the bisection algorithm 


Steps of Bisection Algorithm
<ol>
<li>Choose two values of $x$ between which function has a zero.
 <li> Calculate the function value at both edges of the interval, $f(a)$ and $f(b)$   
<li> Calculate the value of the function at the middle of the interval, $x_h=(a+b)/2$.
<li> if $ f(a) \times f(x_h) < 0 $ root is in between $a$ and $x_h$, then $ b = x_h $, else $a = x_h$. 
<li>  if $abs(f(b)-f(a))/f(a) < \varepsilon$ return $(a+b)/2 $ else Go to step 3
</ol>

```python
from math import cos,sin,fabs,pi

def f(x):
    return x**2 - 1
def df(x):
    return 2*x

x0 = 2.0

def Newton(f,df,x0):
    x =0.0
    for i in range(5):
        x = x0 - f(x0)/df(x0)
        x0 = x
    return(x)

def secant(f,x0,h):
    x =0.0
    for i in range(5):
        x = x0 - 2*h* f(x0)/(f(x0+h)-f(x0-h))
        x0 = x
    return(x)

def Bisection(f,a,b,eps):
    f1 = f(a)
    f2 = f(b)
    if f1*f2 > 0 : 
        print(" Wrong interval")
        return -999
    # while fabs((f2-f1)/f2) > eps:
    for i in range(10000):
        x = (a+b)/2
        f3 = f(x)
        if f1*f3 < 0 :
           f2=f3
           b = x
        else :
           f1=f3
           a = x  
        if(fabs(f1-f2)<eps) : break    
    return (a+b)/2  

print(Newton(f,df,3))
print(secant(f,3,0.1))
print(Bisection(f,-1,2,0.01))
```

# Integration

This section is devoted to the  calculation of the definite integral
of continuous functions: $\int_a^b f(x) dx $, between the limits, $[a,b]\; (a<b)$.  If the analytical integration of a function is difficult, one may employ numerical integration techniques.

In numerical integration, discrete values of a function may be known, at some discrete values of the independent variable $x$. Hence, interpolation plays the most important role in filling the gaps between the discrete points. Knowing a vector of function values ${f_i}$, if a good interpolation can give the function value at any point in the interval $\Delta x$, between the known values of the function, the integral becomes a success. The error in the integration is also closely dependent on the error of representing the function between the mesh points.  The function values may be evenly or unevenly spaced. Here, $N(even)$ equally spaced function value will be regarded.  
Mathematically, the Riemann definition of an integral is the limit,

$$ \int_a^b f(x) \;dx = lim_{h\rightarrow 0} \sum_{i=1}^{(b-a)/h } f(x_i) h $$


Numerical integration is similar, but approximates the integral as the a finite sum
over rectangles of height $f(x_i)$ and widths (or weights) $w_i$

$$ \int_a^b f(x) \;dx = lim_{h\rightarrow 0} \sum_{i=1}^{N} f(x_i) h $$

\begin{equation}
N=\frac{b-a}{\Delta x} 
\end{equation}

 Since the integral may be written as the sum of integrals, it is sufficient to define an interpolation formula for the interval $\Delta x$.


\begin{equation}
\int_a^bf(x)dx=\int_a^{a+\Delta x}f(x)dx+\int_{a+\Delta x}^{a+2\Delta x}f(x)dx+\dots+\int_{b-\Delta x}^b f(x)dx
\end{equation}


If the function is smooth, in these small intervals a
Taylor series expansion at the middle of the interval may be a good
approximation.  By making an appropriate change of variable integrals throughout $\Delta x$ may be written such that, 


$$\int_{x_0-\Delta x}^{x_0+\Delta x}f(x)dx $$
By using difference formulas,the Taylor seies expansion at $x_0$ is,



\begin{multline}
f(x)=f(x_0)+\frac{f(x_0+\Delta x)-f(x_0-\Delta x)}{2 \Delta x}(x-x_0)+ \\ 
\frac{1}{2}\frac{f(x_0+\Delta x)-2f(x_0)+f(x_0-\Delta x)}{\Delta x^2}(x -x_0)^2+\mathcal{O}(\Delta x^3)
\end{multline}


where only the terms up to $\mathcal{O} (\Delta x^3)$ is considered. 
##  Trapezoid Rule

By taking only the first two terms of the series expansion,the integral becomes,


\begin{eqnarray}
\int_{x_0}^{x_0+\Delta x} f(x) \;dx &=& \left .  f(x_0) x \right |_{x_0}^{x_0+\Delta x} + \frac{1}{2} \frac{f(x_0  +\Delta x) - f(x_0)}{\Delta x} \left .(x-x0)^2 \right |_{x_0} ^{x_0+\Delta x}+ \mathcal{O}(\Delta x^3)\\
   &\sim & f(x_0) \Delta x +\frac{1}{2} \frac{f(x_0  + \Delta x) - f(x_0)}{\Delta x} \left[(x_0 + \Delta x - x_0)^2 - (x_0 - x_0)^2\right ]\\
   &\sim & f(x_0) \Delta x +\frac{1}{2} \frac{f(x_0  + \Delta x) - f(x_0)}{\Delta x} {\Delta x }^2 \\
     & \sim &  \frac{\Delta x}{2} \left[  f(x_0)  + f(x_0  + \Delta x) \right]
\end{eqnarray}
 
 
 \begin{equation}
\int_a^bf(x)dx=\int_a^{a+\Delta x}f(x)dx+\int_{a+\Delta x}^{a+2\Delta x}f(x)dx+\dots+\int_{b-\Delta x}^b f(x)dx
\end{equation}


 \begin{equation}
\int_a^bf(x)dx= \frac{\Delta x}{2} \left[  f(a)  + f(a  + \Delta x) + f(a  + \Delta x) + f(a  + 2 \Delta x) \dots + f(b  - \Delta x) + f(b)\right]
\end{equation}

\begin{equation}
\int_a^bf(x)dx= \Delta x \left[\frac{f(a)  + f(b) }{2} +  \left[  f(a  + \Delta x) + f(a  + 2 \Delta x) \dots + f(b  - \Delta x)\right] \right]
\end{equation}

$$ Err_{Trapezoid} = \mathcal{O}\left( \frac{|b-a|^3}{N^2} \right) \frac{d^2f}{dx^2} $$

## Simpson’s Rule

\begin{equation}
\int_a^bf(x)dx=\int_a^{a+\Delta x}f(x)dx+\int_{a+\Delta x}^{a+2\Delta x}f(x)dx+\dots+\int_{b-\Delta x}^b f(x)dx
\end{equation}

Simpson’s rule approximates the integrand $f(x)$ by a parabola , $f(x) \sim \alpha x^2 = \beta x + \gamma $

Simpson’s rule requires the elementary integration to be over pairs of intervals, which in turn
requires that the total number of intervals be even or that the number of points N be odd. In
order to apply Simpson’s rule to the entire interval, we add up the contributions from each pair
of subintervals, counting all but the first and last endpoints twice:

\begin{equation}
\int_a^bf(x)dx=\int_{x_0-\Delta x}^{x_0+\Delta x}f(x)dx+\int_{x_0+\Delta x}^{a+3\Delta x}f(x)dx+\dots+\int_{b-2\Delta x}^b f(x)dx
\end{equation}




$$ \int_{x_0-\Delta x}^{x_0+\Delta x}f(x)dx=2\Delta x f(x_0)+\mathcal{O}(\Delta x^3)$$

since th interval is symmetric around $x_0$, no contribution comes from the second term. Considering all terms
up to the order $\Delta x^3$,

$$ \int_{x_0-\Delta x}^{x_0+\Delta x}f(x)dx=\frac{\Delta x}{3}(f(x_0+\Delta x)+4f(x_0)+f(x_0+\Delta x))+\mathcal{O}(\Delta x^3) 
$$

here since $\mathcal{O}(\Delta x^3)$ term has no contribution error term is at $\mathcal{O}(\Delta x^5)$. This formula is known to be Simpson's rule. Summing up all terms from $a$ to $b$, the integral becomes,

\begin{multline}
\int_a^bf(x)dx = \frac{\Delta x}{3}[f(a)+4f(a+\Delta x)+2f(a+2\Delta x)\\ 
               +4f(a+3\Delta x)+2f(a+4\Delta x)+\dots \\
               +2f(b-2\Delta x)+4f(b-\Delta x)+f(b)]
\end{multline}
 In closed form, Simpson's integral formula becomes,
 
 \begin{multline}
\int_a^bf(x)dx = \frac{\Delta x}{3}\left[ (f(a) + f(b))+ \sum_{n=1}^{N-1}\left [4f(a+(2\times n+1)\Delta x)+2 \times f(a+(2\times n) \Delta x)\right]\right] 
\end{multline}

Simpson's Rule :

$$ w_i = \left \{ \frac{h}{3},\frac{4h}{3},\frac{2h}{3},\frac{4h}{3},\dots, \frac{4h}{3}\right\}$$ 

Because the interpolation is done with sets of three points, the number of points $N$ must be
odd for Simpson’s rule.

Error :

$$ Err_{Simp} = \mathcal{O}\left( \frac{|b-a|^5}{N^4} \right) \frac{d^4f}{dx^4} $$ 


 Special care must be given when the integrand is oscillating or there exist singularities in the integration range.
 
 

```python
from math import pi,sin
N = 80

h = (pi-1)/N   #uniform spacing

x =  [1+i*h for i in range(N)] # np.linspace(1,pi,N)

f = lambda x : sin(x)/(2*x**3)


I = 0.1985572988;
T = 0;         # trapezoidal
S = 0;         # quadrature



# Trapezoidal Rule
#--------------------
for i in range(1,N):
    T = T + f(x[i]);

T = T + 1/2 * (f(x[N-1]) + f(x[0]))
T = T*h;
err_T = abs(I-T)/I * 100;

print(T,err_T)

#  Simpson (quad)
#--------------------
N=10
M = 2 * N + 1
h = (pi-1)/(M-1)   #uniform spacing

x =  [1+i*h for i in range(M)] # np.linspace(1,pi,N)

S = 0
for i in range(0,N):
    S = S + 2*f(x[2*(i+1)]) + 4*f(x[2*i+1]);

S = S + f(x[0]) + f(x[M-1]) 
S = S*h/3       # Simpson integral result
err_S = abs(I-S)/I * 100;


print(T,err_T ,S, err_S,I)
#Display results
```

## Gaussian Quadrature 


When computing integrals of nontrivial functions over general domains, we must resort to numerical methods. 
Numerical integration is also called numerical quadrature. The idea is that the integral is replaced by a sum, 
where the integrand is sampled in a number of discrete points. This can be described as


$$ \int f({x}) d{x} = \sum_i f(x_i) w_i$$

where $x_i$ is the locations of the integration points and $w_i$ is the corresponding weight factors. 
The integration points are often called Gauss points, even though this nomenclature, strictly speaking, 
is correct only for integration points defined by the Gaussian quadrature method. 


### Gaussian Quadrature  (Gauss-Legendre)


In the Gaussian quadrature algorithm, the locations of the integration points and their weights 
are chosen so that a polynomial of as high of a degree as possible can be integrated exactly. 
Since a polynomial of degree $N$ contains $N+1$ coefficients, and a Gauss point rule with $M$ points 
contains $2 M$ parameters (locations and weights), the highest order of the polynomial that can be 
integrated exactly is $N = 2 M - 1$.

Gaussian quadrature is very efficient for integrating fields that can be well approximated by a 
polynomial of a certain degree. The integration points and weights for the first orders of 
Gaussian quadrature in $1-D$ are shown in the table below. The integral is taken over the normalized 
interval [-1,1].



The $N$ point Gauss quadrature integration algorithm gives integration over $[-1,1]$ exactly if $g(x)$ 
is a polynomial of degree $2N - 1$ or less. The $N$ zeros of the Legendre polynomial of degree $N$ are 
the unevenly distributed, and the weights related to the derivatives of the polynomials.

$$ P_N(x_i) = 0, \;\; \;\;\; w_i = \frac{2}{(1-x_i)^2[P_N^{'}]^2)} $$


If the integration range is $[a,b]$, change variables. Since original coordinates are in between $-1 < \; x_i\; < 1$,

$$ x_i^{'} = \frac{b+a}{2} + \frac{b-a}{2} x_i \;\;\;\; w_i^{'} = \frac{b-a}{2} \; w_i $$

Gaussian quadrature will produce higher accuracy than the trapezoid and
Simpson rules for the same number of points.

# Gauss - Laguere  Integration

$$ I = \int_0^{\infty} e^{-x} f(x) dx $$

$$ x y^{\prime\prime} + (1-x)y^{\prime} + n y = 0$$

Solution is $L_n(x)$ where, $ L_0(x) = 1,\;\;L_1(x) = -x+1 ,\;\; L_2(x) = \frac{1}{2} (x^2 -4 x +2),\;\;\dots $

$x_i$'s are the $i^{\rm th}$ root of the laguere polinomial, $L_n(x)$. Using $x_i$, corresponding weights:

$$ w_i = \frac{x_i}{(n+1)^2[L_{n+1}(x_i)]^2} $$


To tke an integral of the form,

$$ \int_0^{\infty} g(x) dx $$ the integral must be rewritten to obtain the standard form,

$$ \int_0^{\infty} g(x) dx  = \int_0^{\infty} g(x) e^x \;e^{-x}dx = \int_0^{\infty} f(x) e^{-x} dx   $$

Here $f(x) = g(x)\; e^x$

## Gauss-Hermite Integration

$$ \int_{-\infty}^{\infty} e^{-x^2}f(x) dx $$

Hermite differential equation,

Solution is $H_n(x)$ Hermite polinomials, $ H_0(x) = 0,\; H_1(x) = 2 x,\;\; H_2(x) = 4 x^2 - 2,\;\; H_3(x) = 8 x^3 - 12 x $,

$$ w_n = \frac{2^{n-1} n! \sqrt{\pi}}{ n^2 \;[H_{n-1}(x_i)]^2} $$

here, $x_i$'s are the roots of $n^{\rm th}$ hermite polinom.
